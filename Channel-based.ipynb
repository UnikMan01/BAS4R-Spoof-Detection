```python
import os, glob, random
import numpy as np
import librosa
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    classification_report, confusion_matrix, accuracy_score,
    roc_curve, auc, precision_recall_curve, average_precision_score
)
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt
from tqdm import tqdm

# ----------------- CONFIG -----------------
ROOT = r"F:\Final Data set\Channel-based\Telephone"

# Folder names (edit if spelling differs)
CLASS_FOLDERS = [
    "Femail Fake",
    "Femail Real",
    "Male Fake",
    "Male Real",
]

SEED = 42
SR = 16000
N_MFCC = 40
HOP_LENGTH = 160          # 10ms hop at 16k
N_FFT = 400               # 25ms window at 16k
FIX_SECONDS = 3.0         # each audio -> 3s (pad/crop). Increase if needed.
FIX_SAMPLES = int(SR * FIX_SECONDS)

BATCH_SIZE = 64
EPOCHS = 30
LR = 1e-3
WEIGHT_DECAY = 1e-4

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", DEVICE)

# ----------------- REPRODUCIBILITY -----------------
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)

# ----------------- INDEX FILES -----------------
def list_audio_files(folder):
    exts = ("*.wav","*.mp3","*.flac","*.m4a","*.ogg","*.opus")
    files = []
    for e in exts:
        files += glob.glob(os.path.join(folder, e))
    return sorted(files)

paths = []
labels = []
for idx, cname in enumerate(CLASS_FOLDERS):
    fdir = os.path.join(ROOT, cname)
    if not os.path.isdir(fdir):
        raise FileNotFoundError(f"Folder not found: {fdir}")
    fs = list_audio_files(fdir)
    paths += fs
    labels += [idx]*len(fs)
    print(cname, "->", len(fs))

paths = np.array(paths)
labels = np.array(labels)

# split (stratified)
train_p, test_p, train_y, test_y = train_test_split(
    paths, labels, test_size=0.2, random_state=SEED, stratify=labels
)

# ----------------- DATASET -----------------
def load_and_fix(path, sr=SR, fix_samples=FIX_SAMPLES):
    y, _ = librosa.load(path, sr=sr, mono=True)
    if len(y) < fix_samples:
        y = np.pad(y, (0, fix_samples-len(y)), mode="constant")
    else:
        # random crop for train handled outside if you want; for now center crop
        start = (len(y) - fix_samples)//2
        y = y[start:start+fix_samples]
    return y

def mfcc_feature(y, sr=SR):
    mfcc = librosa.feature.mfcc(
        y=y, sr=sr, n_mfcc=N_MFCC, n_fft=N_FFT, hop_length=HOP_LENGTH
    )
    # CMVN normalize (per file)
    mfcc = (mfcc - mfcc.mean(axis=1, keepdims=True)) / (mfcc.std(axis=1, keepdims=True) + 1e-9)
    return mfcc.astype(np.float32)  # shape: (n_mfcc, T)

class MFCCDataset(Dataset):
    def __init__(self, paths, labels, train=True):
        self.paths = list(paths)
        self.labels = list(labels)
        self.train = train

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, i):
        p = self.paths[i]
        y = librosa.load(p, sr=SR, mono=True)[0]

        # Train-time random crop/pad for robustness
        if self.train:
            if len(y) < FIX_SAMPLES:
                y = np.pad(y, (0, FIX_SAMPLES-len(y)), mode="constant")
            else:
                start = np.random.randint(0, len(y)-FIX_SAMPLES+1)
                y = y[start:start+FIX_SAMPLES]
        else:
            # center crop
            if len(y) < FIX_SAMPLES:
                y = np.pad(y, (0, FIX_SAMPLES-len(y)), mode="constant")
            else:
                start = (len(y)-FIX_SAMPLES)//2
                y = y[start:start+FIX_SAMPLES]

        x = mfcc_feature(y)  # (C,T)
        x = torch.from_numpy(x).unsqueeze(0)  # (1,C,T) for 1D conv over T
        y = torch.tensor(self.labels[i]).long()
        return x, y

train_ds = MFCCDataset(train_p, train_y, train=True)
test_ds  = MFCCDataset(test_p,  test_y,  train=False)

# Windows
train_dl = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=0, pin_memory=True)
test_dl  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False, num_workers=0, pin_memory=True)

# ----------------- MODEL (Small 1D-CNN, very fast) -----------------
class SmallCNN(nn.Module):
    def __init__(self, n_classes=4):
        super().__init__()
        # input: (B,1,40,T) -> treat 40 as "channels" by first conv2d-like using Conv1d on T after flattening C
        # We'll reshape to (B,40,T) and run Conv1d (fast).
        self.conv1 = nn.Conv1d(N_MFCC, 64, kernel_size=5, stride=1, padding=2)
        self.bn1   = nn.BatchNorm1d(64)
        self.conv2 = nn.Conv1d(64, 128, kernel_size=5, stride=1, padding=2)
        self.bn2   = nn.BatchNorm1d(128)
        self.conv3 = nn.Conv1d(128, 128, kernel_size=3, stride=1, padding=1)
        self.bn3   = nn.BatchNorm1d(128)

        self.pool  = nn.MaxPool1d(2)
        self.drop  = nn.Dropout(0.25)
        self.gap   = nn.AdaptiveAvgPool1d(1)
        self.fc    = nn.Linear(128, n_classes)

    def forward(self, x):
        # x: (B,1,C,T) -> (B,C,T)
        x = x.squeeze(1)

        x = self.pool(F.relu(self.bn1(self.conv1(x))))
        x = self.drop(x)
        x = self.pool(F.relu(self.bn2(self.conv2(x))))
        x = self.drop(x)
        x = self.pool(F.relu(self.bn3(self.conv3(x))))
        x = self.drop(x)

        x = self.gap(x).squeeze(-1)  # (B,128)
        x = self.fc(x)
        return x

model = SmallCNN(n_classes=len(CLASS_FOLDERS)).to(DEVICE)
opt = torch.optim.AdamW(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
criterion = nn.CrossEntropyLoss()

scaler = torch.cuda.amp.GradScaler(enabled=(DEVICE=="cuda"))

# ----------------- TRAIN LOOP -----------------
train_losses, train_accs = [], []
val_losses, val_accs = [], []

def eval_epoch():
    model.eval()
    all_preds, all_true = [], []
    running_loss = 0.0
    with torch.no_grad():
        for xb, yb in test_dl:
            xb, yb = xb.to(DEVICE, non_blocking=True), yb.to(DEVICE, non_blocking=True)
            logits = model(xb)
            loss = criterion(logits, yb)
            running_loss += loss.item() * xb.size(0)
            preds = logits.argmax(dim=1)
            all_preds.append(preds.cpu().numpy())
            all_true.append(yb.cpu().numpy())
    all_preds = np.concatenate(all_preds)
    all_true = np.concatenate(all_true)
    return running_loss/len(test_ds), accuracy_score(all_true, all_preds)

for ep in range(1, EPOCHS+1):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for xb, yb in tqdm(train_dl, desc=f"Epoch {ep}/{EPOCHS}", leave=False):
        xb, yb = xb.to(DEVICE, non_blocking=True), yb.to(DEVICE, non_blocking=True)
        opt.zero_grad(set_to_none=True)

        with torch.cuda.amp.autocast(enabled=(DEVICE=="cuda")):
            logits = model(xb)
            loss = criterion(logits, yb)

        scaler.scale(loss).backward()
        scaler.step(opt)
        scaler.update()

        running_loss += loss.item() * xb.size(0)
        preds = logits.argmax(dim=1)
        correct += (preds == yb).sum().item()
        total += xb.size(0)

    tr_loss = running_loss/len(train_ds)
    tr_acc  = correct/total

    va_loss, va_acc = eval_epoch()

    train_losses.append(tr_loss); train_accs.append(tr_acc)
    val_losses.append(va_loss);   val_accs.append(va_acc)

    print(f"Epoch {ep:02d} | train loss {tr_loss:.4f} acc {tr_acc:.4f} | val loss {va_loss:.4f} acc {va_acc:.4f}")

# ----------------- FINAL EVAL (Report + Confusion) -----------------
model.eval()
all_probs = []
all_preds = []
all_true  = []

with torch.no_grad():
    for xb, yb in test_dl:
        xb = xb.to(DEVICE, non_blocking=True)
        logits = model(xb)
        probs = F.softmax(logits, dim=1).cpu().numpy()
        preds = probs.argmax(axis=1)
        all_probs.append(probs)
        all_preds.append(preds)
        all_true.append(yb.numpy())

all_probs = np.vstack(all_probs)
all_preds = np.concatenate(all_preds)
all_true  = np.concatenate(all_true)

print("\nAccuracy:", accuracy_score(all_true, all_preds))
print("\nClassification Report:\n")
print(classification_report(all_true, all_preds, target_names=CLASS_FOLDERS))

cm = confusion_matrix(all_true, all_preds)
print("Confusion Matrix:\n", cm)

# ----------------- PLOTS: Loss / Acc curves -----------------
plt.figure()
plt.plot(train_losses, label="train_loss")
plt.plot(val_losses, label="val_loss")
plt.title("Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()

plt.figure()
plt.plot(train_accs, label="train_acc")
plt.plot(val_accs, label="val_acc")
plt.title("Accuracy Curve")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

# ----------------- Confusion Matrix Plot -----------------
plt.figure()
plt.imshow(cm, interpolation="nearest")
plt.title("Confusion Matrix")
plt.colorbar()
tick_marks = np.arange(len(CLASS_FOLDERS))
plt.xticks(tick_marks, CLASS_FOLDERS, rotation=45, ha="right")
plt.yticks(tick_marks, CLASS_FOLDERS)
plt.xlabel("Predicted")
plt.ylabel("True")

# write values
thresh = cm.max() / 2.0 if cm.size else 0.0
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, format(cm[i, j], "d"),
                 ha="center", va="center")
plt.tight_layout()
plt.show()

# ----------------- ROC + PR (One-vs-Rest) -----------------
y_true_bin = label_binarize(all_true, classes=list(range(len(CLASS_FOLDERS))))

# ROC
plt.figure()
for i, name in enumerate(CLASS_FOLDERS):
    fpr, tpr, _ = roc_curve(y_true_bin[:, i], all_probs[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f"{name} (AUC={roc_auc:.3f})")
plt.plot([0,1], [0,1], linestyle="--")
plt.title("ROC Curve (OvR)")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()

# PR
plt.figure()
for i, name in enumerate(CLASS_FOLDERS):
    prec, rec, _ = precision_recall_curve(y_true_bin[:, i], all_probs[:, i])
    ap = average_precision_score(y_true_bin[:, i], all_probs[:, i])
    plt.plot(rec, prec, label=f"{name} (AP={ap:.3f})")
plt.title("Precision-Recall Curve (OvR)")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.legend()
plt.show()
```

```python
import os, glob, random
import numpy as np
import librosa
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    classification_report, confusion_matrix, accuracy_score,
    roc_curve, auc, precision_recall_curve, average_precision_score
)
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt
from tqdm import tqdm

# ----------------- CONFIG -----------------
ROOT = r"F:\Final Data set\Channel-based\VoIP Low-Bitrate"

# Folder names (edit if spelling differs)
CLASS_FOLDERS = [
    "Femail Fake",
    "Femail Real",
    "Male Fake",
    "Male Real",
]

SEED = 42
SR = 16000
N_MFCC = 40
HOP_LENGTH = 160          # 10ms hop at 16k
N_FFT = 400               # 25ms window at 16k
FIX_SECONDS = 3.0         # each audio -> 3s (pad/crop). Increase if needed.
FIX_SAMPLES = int(SR * FIX_SECONDS)

BATCH_SIZE = 64
EPOCHS = 30
LR = 1e-3
WEIGHT_DECAY = 1e-4

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", DEVICE)

# ----------------- REPRODUCIBILITY -----------------
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)

# ----------------- INDEX FILES -----------------
def list_audio_files(folder):
    exts = ("*.wav","*.mp3","*.flac","*.m4a","*.ogg","*.opus")
    files = []
    for e in exts:
        files += glob.glob(os.path.join(folder, e))
    return sorted(files)

paths = []
labels = []
for idx, cname in enumerate(CLASS_FOLDERS):
    fdir = os.path.join(ROOT, cname)
    if not os.path.isdir(fdir):
        raise FileNotFoundError(f"Folder not found: {fdir}")
    fs = list_audio_files(fdir)
    paths += fs
    labels += [idx]*len(fs)
    print(cname, "->", len(fs))

paths = np.array(paths)
labels = np.array(labels)

# split (stratified)
train_p, test_p, train_y, test_y = train_test_split(
    paths, labels, test_size=0.2, random_state=SEED, stratify=labels
)

# ----------------- DATASET -----------------
def load_and_fix(path, sr=SR, fix_samples=FIX_SAMPLES):
    y, _ = librosa.load(path, sr=sr, mono=True)
    if len(y) < fix_samples:
        y = np.pad(y, (0, fix_samples-len(y)), mode="constant")
    else:
        # random crop for train handled outside if you want; for now center crop
        start = (len(y) - fix_samples)//2
        y = y[start:start+fix_samples]
    return y

def mfcc_feature(y, sr=SR):
    mfcc = librosa.feature.mfcc(
        y=y, sr=sr, n_mfcc=N_MFCC, n_fft=N_FFT, hop_length=HOP_LENGTH
    )
    # CMVN normalize (per file)
    mfcc = (mfcc - mfcc.mean(axis=1, keepdims=True)) / (mfcc.std(axis=1, keepdims=True) + 1e-9)
    return mfcc.astype(np.float32)  # shape: (n_mfcc, T)

class MFCCDataset(Dataset):
    def __init__(self, paths, labels, train=True):
        self.paths = list(paths)
        self.labels = list(labels)
        self.train = train

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, i):
        p = self.paths[i]
        y = librosa.load(p, sr=SR, mono=True)[0]

        # Train-time random crop/pad for robustness
        if self.train:
            if len(y) < FIX_SAMPLES:
                y = np.pad(y, (0, FIX_SAMPLES-len(y)), mode="constant")
            else:
                start = np.random.randint(0, len(y)-FIX_SAMPLES+1)
                y = y[start:start+FIX_SAMPLES]
        else:
            # center crop
            if len(y) < FIX_SAMPLES:
                y = np.pad(y, (0, FIX_SAMPLES-len(y)), mode="constant")
            else:
                start = (len(y)-FIX_SAMPLES)//2
                y = y[start:start+FIX_SAMPLES]

        x = mfcc_feature(y)  # (C,T)
        x = torch.from_numpy(x).unsqueeze(0)  # (1,C,T) for 1D conv over T
        y = torch.tensor(self.labels[i]).long()
        return x, y

train_ds = MFCCDataset(train_p, train_y, train=True)
test_ds  = MFCCDataset(test_p,  test_y,  train=False)

# Windows 
train_dl = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=0, pin_memory=True)
test_dl  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False, num_workers=0, pin_memory=True)

# ----------------- MODEL (Small 1D-CNN, very fast) -----------------
class SmallCNN(nn.Module):
    def __init__(self, n_classes=4):
        super().__init__()
        # input: (B,1,40,T) -> treat 40 as "channels" by first conv2d-like using Conv1d on T after flattening C
        # We'll reshape to (B,40,T) and run Conv1d (fast).
        self.conv1 = nn.Conv1d(N_MFCC, 64, kernel_size=5, stride=1, padding=2)
        self.bn1   = nn.BatchNorm1d(64)
        self.conv2 = nn.Conv1d(64, 128, kernel_size=5, stride=1, padding=2)
        self.bn2   = nn.BatchNorm1d(128)
        self.conv3 = nn.Conv1d(128, 128, kernel_size=3, stride=1, padding=1)
        self.bn3   = nn.BatchNorm1d(128)

        self.pool  = nn.MaxPool1d(2)
        self.drop  = nn.Dropout(0.25)
        self.gap   = nn.AdaptiveAvgPool1d(1)
        self.fc    = nn.Linear(128, n_classes)

    def forward(self, x):
        # x: (B,1,C,T) -> (B,C,T)
        x = x.squeeze(1)

        x = self.pool(F.relu(self.bn1(self.conv1(x))))
        x = self.drop(x)
        x = self.pool(F.relu(self.bn2(self.conv2(x))))
        x = self.drop(x)
        x = self.pool(F.relu(self.bn3(self.conv3(x))))
        x = self.drop(x)

        x = self.gap(x).squeeze(-1)  # (B,128)
        x = self.fc(x)
        return x

model = SmallCNN(n_classes=len(CLASS_FOLDERS)).to(DEVICE)
opt = torch.optim.AdamW(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
criterion = nn.CrossEntropyLoss()

scaler = torch.cuda.amp.GradScaler(enabled=(DEVICE=="cuda"))

# ----------------- TRAIN LOOP -----------------
train_losses, train_accs = [], []
val_losses, val_accs = [], []

def eval_epoch():
    model.eval()
    all_preds, all_true = [], []
    running_loss = 0.0
    with torch.no_grad():
        for xb, yb in test_dl:
            xb, yb = xb.to(DEVICE, non_blocking=True), yb.to(DEVICE, non_blocking=True)
            logits = model(xb)
            loss = criterion(logits, yb)
            running_loss += loss.item() * xb.size(0)
            preds = logits.argmax(dim=1)
            all_preds.append(preds.cpu().numpy())
            all_true.append(yb.cpu().numpy())
    all_preds = np.concatenate(all_preds)
    all_true = np.concatenate(all_true)
    return running_loss/len(test_ds), accuracy_score(all_true, all_preds)

for ep in range(1, EPOCHS+1):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for xb, yb in tqdm(train_dl, desc=f"Epoch {ep}/{EPOCHS}", leave=False):
        xb, yb = xb.to(DEVICE, non_blocking=True), yb.to(DEVICE, non_blocking=True)
        opt.zero_grad(set_to_none=True)

        with torch.cuda.amp.autocast(enabled=(DEVICE=="cuda")):
            logits = model(xb)
            loss = criterion(logits, yb)

        scaler.scale(loss).backward()
        scaler.step(opt)
        scaler.update()

        running_loss += loss.item() * xb.size(0)
        preds = logits.argmax(dim=1)
        correct += (preds == yb).sum().item()
        total += xb.size(0)

    tr_loss = running_loss/len(train_ds)
    tr_acc  = correct/total

    va_loss, va_acc = eval_epoch()

    train_losses.append(tr_loss); train_accs.append(tr_acc)
    val_losses.append(va_loss);   val_accs.append(va_acc)

    print(f"Epoch {ep:02d} | train loss {tr_loss:.4f} acc {tr_acc:.4f} | val loss {va_loss:.4f} acc {va_acc:.4f}")

# ----------------- FINAL EVAL (Report + Confusion) -----------------
model.eval()
all_probs = []
all_preds = []
all_true  = []

with torch.no_grad():
    for xb, yb in test_dl:
        xb = xb.to(DEVICE, non_blocking=True)
        logits = model(xb)
        probs = F.softmax(logits, dim=1).cpu().numpy()
        preds = probs.argmax(axis=1)
        all_probs.append(probs)
        all_preds.append(preds)
        all_true.append(yb.numpy())

all_probs = np.vstack(all_probs)
all_preds = np.concatenate(all_preds)
all_true  = np.concatenate(all_true)

print("\nAccuracy:", accuracy_score(all_true, all_preds))
print("\nClassification Report:\n")
print(classification_report(all_true, all_preds, target_names=CLASS_FOLDERS))

cm = confusion_matrix(all_true, all_preds)
print("Confusion Matrix:\n", cm)

# ----------------- PLOTS: Loss / Acc curves -----------------
plt.figure()
plt.plot(train_losses, label="train_loss")
plt.plot(val_losses, label="val_loss")
plt.title("Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()

plt.figure()
plt.plot(train_accs, label="train_acc")
plt.plot(val_accs, label="val_acc")
plt.title("Accuracy Curve")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

# ----------------- Confusion Matrix Plot -----------------
plt.figure()
plt.imshow(cm, interpolation="nearest")
plt.title("Confusion Matrix")
plt.colorbar()
tick_marks = np.arange(len(CLASS_FOLDERS))
plt.xticks(tick_marks, CLASS_FOLDERS, rotation=45, ha="right")
plt.yticks(tick_marks, CLASS_FOLDERS)
plt.xlabel("Predicted")
plt.ylabel("True")

# write values
thresh = cm.max() / 2.0 if cm.size else 0.0
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, format(cm[i, j], "d"),
                 ha="center", va="center")
plt.tight_layout()
plt.show()

# ----------------- ROC + PR (One-vs-Rest) -----------------
y_true_bin = label_binarize(all_true, classes=list(range(len(CLASS_FOLDERS))))

# ROC
plt.figure()
for i, name in enumerate(CLASS_FOLDERS):
    fpr, tpr, _ = roc_curve(y_true_bin[:, i], all_probs[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f"{name} (AUC={roc_auc:.3f})")
plt.plot([0,1], [0,1], linestyle="--")
plt.title("ROC Curve (OvR)")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()

# PR
plt.figure()
for i, name in enumerate(CLASS_FOLDERS):
    prec, rec, _ = precision_recall_curve(y_true_bin[:, i], all_probs[:, i])
    ap = average_precision_score(y_true_bin[:, i], all_probs[:, i])
    plt.plot(rec, prec, label=f"{name} (AP={ap:.3f})")
plt.title("Precision-Recall Curve (OvR)")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.legend()
plt.show()
```
